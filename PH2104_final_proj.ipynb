# Tidal field at a planet

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('_mpl-gallery-nogrid')
```

```python
x = np.linspace(-1, 1, 15)
y = np.linspace(-1, 1, 15)
X, Y = np.meshgrid(x, y)
U = 1-10000*(100+X)/(((100+X)**2+Y**2)**(3/2))
V = -10000*Y/(((100+X)**2+Y**2)**(3/2))

planet = plt.Circle((0, 0), 0.6, alpha=0.6)

fig, ax = plt.subplots(figsize=(5, 5))
ax.quiver(X, Y, U, V)
ax.add_patch(planet)

plt.show()
```

# Sea surface deformation

```python
theta = np.arange(0, np.pi, 0.1)
h = 1*(10+1)*(np.sin(theta)**2)

plt.figure(figsize=(6, 2.5))
plt.plot(theta, h)
plt.text(0.1, 2, r'$\gamma=10$', fontsize=10)
plt.title(r'$h(\theta)-h_0$', fontsize=15)
plt.xlabel(r'$\theta$ [rad]')
plt.xlim(0, np.pi)
plt.ylim(0, 12)
plt.show()
```

```python
t = np.arange(0, 1, 0.01)
h = 1*(10 + np.cos(2*np.pi*t)**2)*1

plt.figure(figsize=(6, 2.5))
plt.plot(t, h)
plt.text(0.1, 2, r'$\gamma=10$', fontsize=10)
plt.title(r'$h(\theta=\frac{\pi}{2},t)-h_0$', fontsize=15)
plt.xlabel(r'$t$ [T]')
plt.xlim(0, 1)
plt.ylim(0, 12)
plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(projection='3d')

# # Grab some test data.
# X, Y, Z = axes3d.get_test_data(0.05)

theta = np.arange(0, np.pi, np.pi/100)
t = np.arange(0, 1, 0.01)
# h = 1*(10+1)*(np.sin(theta)**2)
# h = 1*(10 + np.cos(2*np.pi*t)**2)*1
# h = 1*(10 + np.cos(2*np.pi*t)**2)*(np.sin(theta)**2)

Theta, T = np.meshgrid(theta, t)

H = 1*(10 + np.cos(2*np.pi*T)**2)*(np.sin(Theta)**2)

# Plot a basic wireframe.
ax.plot_wireframe(Theta, T, H, rstride=5, cstride=5)

plt.xlabel(r'$\theta$ [rad]')
plt.ylabel(r'$t$ [T]')
plt.title(r'$h(\theta,t)-h_0$', fontsize=15)

ax.set_xlim(0, np.pi)
ax.set_ylim(0, 1)
ax.set_zlim(0, 12)

ax.text(0, 0, 2, r'$\gamma=10$', fontsize=10)

plt.show()
```
